```python
# importing packages
import json
import pandas as pd
from pandasql import sqldf
import pandasql as ps
```

```python
# Converting Brands data from json file
df1 = open(r'brands.json')
brands = [json.loads(line) for line in df1]

# Converting users data from json file
df2 = open(r'users.json')
users = [json.loads(line) for line in df2]
    
# Converting receipts data from json file    
df3 = open(r'receipts.json')
receipts = [json.loads(line) for line in df3]
```

```python
# Normalizaing the data
df_brands = pd.json_normalize(brands)
df_users = pd.json_normalize(users)
df_receipts = pd.json_normalize(receipts)
```
